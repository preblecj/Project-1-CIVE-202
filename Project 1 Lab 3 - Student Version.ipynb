```python
### Sample temperature function

#step 1 - define function that assigns temperature levels

def assign_temperature(temperature):
    if temperature < 32:
        return 'Below Freezing'
    elif 32 <= temperature <= 50:
        return 'Cool'
    elif 51 <= temperature <= 70:
        return 'Warm'
    else:
        return 'Hot' 

#step 2 - apply the function to data


```

Now that we have the function, we need to `apply` that function to our data. If we have a created a function, it doesn't mean this function has been used yet, its just defined in what we call the 'global environment'.  To make sure our function is applied to our data our steps are as follows:
1. Make sure we `apply` the function to the column we want to be manipulated
2. Assign the new column we create to a column name inside our original dataset

```python
## Apply your function to the temperature column in your air_data

#step 2 - apply function

#air_data['temperature_level'] #this creates a new column inside air_data called temperature level
air_data['temperature_level'] = air_data['temperature'].apply(assign_temperature) 
#the input to my assign_temperature function would be the data coulmn temperature

air_data.head(5)
```

#### Lab Activity #1: Create a function for altitude levels
Create a function that assigns three different levels to altitude: Low, Medium and High.  You can choose what values to set as your conditions in this function.  

```python
### Enter your response to Lab Activity #1 in the space below:

def assign_altitude(altitude):
    if altitude < 1000:
        return 'Low'
    elif 1001 <= altitude <= 2500:
        return 'Medium'
    else:
        return 'High' 

air_data['altitude_height'] = air_data['sensor.altitude'].apply(assign_altitude)

air_data.head(5)
```

### Sorting our data

Let's start by reminding ourselves how to aggregate data.  Let's calculate the max, min, mean, and median of the `pm_2.5` values and name this new data set something besides `air_data`

```python
## Group your data

### Aggregate the grouped data

### Make sure the students take the mean, median and max of the values of the columns.  
## Summary stats on PM 2.5

air_data_grouped = air_data.groupby(['sensor.name'])

air_data_pm25 = air_data_grouped.agg(max=('pm2.5_atm', 'max'),
                                                     min=('pm2.5_atm', 'min'),
                                                     mean=('pm2.5_atm', 'mean'),
                                                     median=('pm2.5_atm', 'median'))

air_data_pm25
```

Now that we have the aggregated data, let's sort our values to find the top 5 highest means by location.  We create a new dataframe and append the `sort_values()` function, giving it a specific set of arguments to let `.sort_values()` know what to sort `by` and if the order is `ascending` or `descending`.

```python
### Find the top 5 locations with highest mean and median values of each parameter

### PM 2.5 Mean

top5_mean_pm25 = (
    air_data_pm25.sort_values(by = 'mean', ascending = False)
)

top5_mean_pm25.head(5)
## Here is a link to how this code was generated: https://chatgpt.com/share/695e9143-2c74-8000-bed2-06b976fb8e95 
```

#### Lab Activity #2: VOC summary
Create an aggregated summary of vocs by the max value.  Print out the first 10 rows of the data set grouped by monitor_index

```python
### Enter your response to Lab Activity #2 in the space below
air_data_grouped = air_data.groupby(['monitor_index'])

air_data_voc = air_data_grouped.agg(max=('voc', 'max'),
                                                     min=('voc', 'min'),
                                                     mean=('voc', 'mean'),
                                                     median=('voc', 'median'))

top10_max_voc = (
    air_data_voc.sort_values(by = 'max', ascending = False)
)

top10_max_voc.head(10)

```

#### Sorting by Dates
Dates can sometimes be difficult to wrangle in Python.  Luckily the AirPurple dataset is a 'clean' dataset which has dates formatted as mm/dd/yy which Python easily recognizes once the `air_data` is in a dataframe.  Let's use `.sort_values()` as above to sort by date.  First we will select only three columns: `date`, `sensor.name` and `voc`.  Then sort by date.

```python
### Apply filters

#Step 1 - only care about data that are VOC > 25

unhealthy_voc = air_data[air_data['voc']>=25]

#Step 2 - sort by date and sensor name

unhealthy_events = unhealthy_voc[
['date', 'sensor.name', 'voc']
].sort_values('date')

unhealthy_events.head(10)

```
