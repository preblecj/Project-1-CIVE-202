```python
##assigns temperature and groups it with temperature
def assign_temperature(temperature):
    if temperature < 32:
        return 'Below Freezing'
    elif 32 <= temperature <= 50:
        return 'Cool'
    elif 51 <= temperature <= 70:
        return 'Warm'
    else:
        return 'Hot' 

air_data['temperature_level'] = air_data['temperature'].apply(assign_temperature)

temperature_by_sensor = (air_data.groupby(['sensor.name', 'temperature_level']).size())
temperature_by_sensor
```

```python
##assign humidity level and groups it with sensor.name
def assign_humidity(humidity):
    if humidity < 50:
        return 'Low'
    elif 51 <= humidity <= 80:
        return 'High'
    elif 80 <= humidity <= 100:
        return 'Very High'
        
air_data['humidity_level'] = air_data['humidity'].apply(assign_humidity)

humidity_by_sensor = (air_data.groupby(['sensor.name', 'humidity_level']).size())
humidity_by_sensor
```

```python
##pm2.5 and AQI's EPA index level warnings
def assign_pm25(pm25):
    if pm25 < 12:
        return 'Good'
    elif 12 <= pm25 <= 35.4:
        return 'Moderate'
    elif 35.5 <= pm25 <= 55.4:
        return 'Unhealthy for Sensitive Groups'
    elif 55.5 <= pm25 <= 150.4:
        return 'Unhealthy'
    elif 150.5 <= pm25 <= 250.4:
        return 'Very Unhealthy'
    else:
        return 'Hazardous'
air_data['pm2.5_aqi'] = air_data['pm2.5_atm'].apply(assign_pm25)
air_data

pm25_sensor = air_data.groupby(['sensor.name', 'pm2.5_aqi']).size()
pm25_sensor
```

```python
##pm10 and AQI's EPA index level warnings
def assign_pm10(pm10):
    if pm10 < 54:
        return 'Good'
    elif 55 <= pm10 <= 154:
        return 'Moderate'
    elif 155 <= pm10 <= 254:
        return 'Unhealthy for Sensitive Groups'
    elif 255 <= pm10 <= 354:
        return 'Unhealthy'
    elif 355 <= pm10 <= 425:
        return 'Very Unhealthy'
    else:
        return 'Hazardous'
air_data['pm10.0_aqi'] = air_data['pm10.0_atm'].apply(assign_pm10)
air_data
pm25_sensor = air_data.groupby(['sensor.name', 'pm10.0_aqi']).size()
pm25_sensor
```
